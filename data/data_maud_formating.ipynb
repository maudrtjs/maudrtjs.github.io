```python
import pandas as pd
import json
import numpy as np
import os
```

```python
data_raw = pd.read_csv('bdd.csv')
```

```python
data_raw.head()
```

```python
data_raw.rename(columns={data_raw.columns[0]:"pk"},inplace=True)
data_raw.rename(columns={"nb_opus":"num_opus"},inplace=True)
```

```python
data_raw.head()
```

```python
data_raw["composition_date"] = pd.to_numeric(data_raw["composition_date"],errors='coerce') 
```

```python
list_raw = []
i=0
for index in (data_raw.set_index('pk')).index:
    content = data_raw.set_index('pk').loc[index].to_dict()
    dict_temp = {"model":"search.sheet",
                 "pk":i}
    dict_temp.update({"fields":content})
    list_raw = list_raw + [dict_temp]
    i+=1
    
```

```python
with open('bdd_json.json', 'w') as fp:
    json.dump(list_raw, fp)
```
